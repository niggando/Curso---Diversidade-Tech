### Series

O objeto fundamental do Pandas são as **Series**, uma classe do pandas.

As Series são as **colunas das tabelas** (que veremos mais a frente), e por baixo dos panos, os dados ficam armazenados como **numpy arrays**!

A diferença é que a série possui um **índice associado**, permitindo o acesso aos conteúdos dessa estrutura por ele, como um dicionário.

Além disso, as séries têm métodos específicos além dos que vimos pra arrays, o que será super útil!

Podemos criar uma série **a partir de uma lista**, usando a função do pandas `pd.Series()`: 

````python
pd.np.array([1, 2, 3])
````

````python
# Criando uma lista de valores
lista = [4, 6, 3, 7, 13]
print("lista", lista)
# Criando um array a partir dessa lista
arr = np.array(lista)
print("arr", arr)
````

````python
# Criando uma série
serie = pd.Series(arr)
print(type(serie))
serie
````

````python
# Pegando os valores, note que é retornando um numpy array
serie.values
````

Crie uma série cujos valores variem de 0 até k.

O indice da série será o valor de k*2 e o seu valor k^2 (k ao quadrado).

Por exemplo:
```
k  idx valor
0   0    0
1   2    1
2   4    4
...
10  20   100
```

Imprima na tela os 3 primeiros valores da serie (head) e os 3 últimos (tail).

Além disso imprima os valores da série e os seus indices

Utilize o loc ou iloc para selecionar os valores de k=3 entre k=6

````python
# resolução do professor

k = 10
dc = {x*2: x**2 for x in range(0, k+1)}

series_drops2 = pd.Series(dc)
display(series_drops2)
print('Head')
display(series_drops2.head(3)) # três primeiros valores
print('Tail')
display(series_drops2.tail(3)) # três últimos valores
print('Valores')
display(series_drops2.values) # Valores da série
print('Index')
display(series_drops2.index.tolist()) # Indices da série

print('<loc> Inclusvo')
print(series_drops2.iloc[3:7])

print('<iloc> Exclusivo')
print(series_drops2.loc[3:6])
````

````python
k = 10
indices = np.arange(0, k+1) * 2
valores = np.arange(0, k+1) ** 2

serie_drops2 = pd.Series(data = valores, index = indices)

print(serie_drops2.head(3)) # três primeiros valores
print(serie_drops2.tail(3)) # três últimos valores
print(serie_drops2.values) # Valores da série
print(serie_drops2.index.tolist()) # Indices da série

print(serie_drops2.iloc[3:7])
    
````
