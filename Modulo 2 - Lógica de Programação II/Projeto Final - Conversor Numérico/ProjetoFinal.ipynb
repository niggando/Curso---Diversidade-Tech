```python
# Dicionários para acesso durante a troca de números

dict_numbers = {
    'um': 1,
    'dois': 2,
    'tres': 3,
    'quatro': 4,
    'cinco': 5,
    'seis': 6,
    'sete': 7,
    'oito': 8,
    'nove': 9,
    'dez': 10,
    'onze': 11,
    'doze': 12,
    'treze': 13,
    'quatorze': 14,
    'quinze': 15,
    'dezesseis': 16,
    'dezessete': 17,
    'dezoito': 18,
    'dezenove': 19,
    'vinte': 20,
    'trinta': 30,
    'quarenta': 40,
    'cinquenta': 50,
    'sessenta': 60,
    'setenta': 70,
    'oitenta': 80,
    'noventa': 90,
    'cem': 100,
    'cento': 100,
    'duzentos': 200,
    'trezentos': 300,
    'quatrocentos': 400,
    'quinhentos': 500,
    'seiscentos': 600,
    'setecentos': 700,
    'oitocentos': 800,
    'novecentos': 900,
}

dict_unity = {
    'mil': 1000,
    'milhao': 1000000,
    'milhoes': 1000000,
    'bilhao': 1000000000,
    'bilhoes': 1000000000
}

dict_char = {
    'k': 1000,
    'm': 1000000,
    'g': 1000000000
}
```

```python
# Funções
# Captura de entradas e tratamento das mesmas
import re

def normalize_text(text):
    text = text.lower().replace(',','.').replace('\n','')
    text = re.sub('[áàäã]','a',text)
    text = re.sub('[éèêẽ]','e',text)
    text = re.sub('[íìï]','i',text)
    text = re.sub('[óòõ]','o',text)
    text = re.sub('[ùúü]','u',text)
    return text


# Função que converte sentença com x.x (k, m ou g)
def convert_number_to_text(txt):
    # Tratamento do texto para termos multiplicadores unidos ao número.
    txt = re.sub('[k]', ' k', txt)
    txt = re.sub('[m]', ' m', txt)
    txt = re.sub('[g]', ' g', txt)
    # Procurando na linha o padrão de texto começado em número
    number = float(re.search('\d*.\d*', txt).group(0))
    # Procurando a letra referente ao numero de casas (k, m, g)
    char = re.search('[a-z]', txt).group(0)
    char = dict_char[char]
    return str(number*char)

# Função que converte texto por extenso para número
def convert_text_to_number(txt):
    # Variável que irá receber o valor final
    number_final = 0
    # Variáveis auxiliares para somar valores divididos por mil/milhar/milhão
    sum = 0
    number = 0
    txt = txt.split(' ')
    for word in txt:
        if word in dict_numbers.keys():
            sum += dict_numbers[word]
            number += sum

        elif word in dict_unity.keys():
            # If para resolver caso mil seja o primeiro termo da str ou o próximo termo após uma unidade
            if word == 'mil' and number == 0:
                number = 1
            number *= dict_unity[word]
            number_final += number
            # Resetando variáveis auxiliares
            number = 0
        sum = 0
    # Última iteração fora do for por ter optado pelo for each, adicionando os números sem unidade
    number_final += number
    return str(number_final)

# Função principal que efetua a leitura do arquivo
def main_function():
    arquivo = open('input.txt', 'r', encoding = 'utf-8')
    output = open('output.txt', 'w', encoding = 'utf-8')
    for linha in arquivo:
        if re.match('\d',linha):
            output.writelines(convert_number_to_text(normalize_text(linha))+'\n')
        else:
            output.writelines(convert_text_to_number(normalize_text(linha))+'\n')

# Chamando a função principal
main_function()
```
